```python
query = {
    "beds": {"$gt": 2},  # 大于 2 张床
    "neighbourhood": "Brooklyn, New York, United States",  # 指定的社区
}


projection = {
    "name": 1, 
    "beds": 1, 
    "review_scores_rating": 1, 
    "price": 1,
    "_id": 0  
}

results = collection.find(query, projection).sort("review_scores_rating", pymongo.DESCENDING)

for result in results:
    print(result)
```
